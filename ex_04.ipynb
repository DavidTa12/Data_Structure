```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pad
import seaborn as sns
```

```python
data = pad.read_csv('data.csv')
display(data)
```

```python
my_dataset_gb = data.groupby(['reg_name', 'year'], as_index=False).sum('amount')
display(my_dataset_gb)hue="sex"
```

```python
ax = sns.lineplot(x = 'year', y ='amount', data=my_dataset_gb, hue="reg_name")
sns.set(rc={'figure.figsize':(24,20)})
```

```python
my_dataset_gb = data.groupby(['tax_name'], as_index=False).sum('amount').sort_values(by=['amount'], ascending=False).head(10)
my_dataset_gb.drop('year', axis=1, inplace=True)
display(my_dataset_gb)
```

```python
g = sns.catplot(
    data=my_dataset_gb, kind="bar",
    x="amount", y="tax_name",
    ci="sd", palette="dark", alpha=.6, height=6
)
```
